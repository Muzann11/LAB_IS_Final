```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
category.value_counts().head(20).plot(kind='bar')
plt.title('Top 20 Job Categories')
plt.xlabel('Job Category')
plt.ylabel('Frequency')
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_category_counts = scaler.fit_transform(category.value_counts().values.reshape(-1, 1))
print(scaled_category_counts)
```
